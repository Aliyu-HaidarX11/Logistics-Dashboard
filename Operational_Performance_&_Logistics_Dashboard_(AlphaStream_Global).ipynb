```python
import pandas as pd
import numpy as np

# Set seed for consistency
np.random.seed(7)

# Generate 500 Shipping Projects (Voyages)
n_projects = 500

data = {
    'Project_ID': [f'AS-2026-{i}' for i in range(n_projects)],
    'Route': np.random.choice(['Lagos-Hamburg', 'Singapore-Rotterdam', 'New York-Antwerp'], n_projects),
    'Budget_Allocated_EUR': np.random.randint(50000, 200000, n_projects),
    'Actual_Spend_EUR': np.random.randint(45000, 220000, n_projects),
    'Planned_Duration_Days': np.random.randint(15, 45, n_projects),
    'Actual_Duration_Days': np.random.randint(14, 55, n_projects),
    'Team_Lead': np.random.choice(['Müller', 'Adeyemi', 'Schmidt', 'Okonkwo'], n_projects),
    'Status': np.random.choice(['Completed', 'In Progress', 'Delayed'], n_projects, p=[0.7, 0.2, 0.1])
}

df_pm = pd.DataFrame(data)

# Create a 'Success' Metric: Budget vs Actual
df_pm['Over_Budget'] = df_pm['Actual_Spend_EUR'] > df_pm['Budget_Allocated_EUR']
df_pm['Delay_Days'] = df_pm['Actual_Duration_Days'] - df_pm['Planned_Duration_Days']

df_pm.to_csv('alphastream_pm_data.csv', index=False)
print("Project 5 Operational Data Generated!")
df_pm.head()
```

## Phase 1: Operational KPI Analysis (AlphaStream Global)
In this section, we calculate key performance indicators (KPIs) to evaluate the efficiency of our logistics routes and the performance of our project leads.

```python
# 1. Calculate Budget Variance % (How much over/under budget are we?)
df_pm['Budget_Variance_Pct'] = ((df_pm['Actual_Spend_EUR'] - df_pm['Budget_Allocated_EUR']) / df_pm['Budget_Allocated_EUR']) * 100

# 2. Calculate Delay Percentage (Actual vs Planned Duration)
df_pm['Delay_Pct'] = ((df_pm['Actual_Duration_Days'] - df_pm['Planned_Duration_Days']) / df_pm['Planned_Duration_Days']) * 100

# 3. Create a 'Financial Health' Column
def get_health(variance):
    if variance < 0: return 'Under Budget (Good)'
    elif variance < 10: return 'On Track'
    else: return 'Over Budget (Action Required)'

df_pm['Financial_Status'] = df_pm['Budget_Variance_Pct'].apply(get_health)

# Show the new operational metrics
df_pm[['Project_ID', 'Route', 'Budget_Variance_Pct', 'Financial_Status']].head()
```

```python
# Group by Route to see where the bottlenecks are
route_summary = df_pm.groupby('Route').agg({
    'Budget_Variance_Pct': 'mean',
    'Delay_Days': 'mean',
    'Project_ID': 'count'
}).rename(columns={'Project_ID': 'Total_Voyages'})

print("--- Executive Route Performance ---")
print(route_summary.sort_values(by='Delay_Days', ascending=False))

# Group by Team Lead to see who manages the budget best
lead_summary = df_pm.groupby('Team_Lead')['Budget_Variance_Pct'].mean().sort_values()

print("\n--- Team Lead Budget Performance (Lower is Better) ---")
print(lead_summary)
```
